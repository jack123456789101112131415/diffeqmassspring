```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

# Parameters defining the system
c = 4  # Damping constant
k = 2  # Stiffness of the spring
m = 20  # Mass
F = 5  # Force

# Simulation Parameters
tstart = 0
tstop = 60
increment = 0.1
t = np.arange(tstart, tstop + increment, increment)

# Force input
Ft = np.ones_like(t) * F

# System matrices
A = [[0, 1], [-k / m, -c / m]]
B = [[0], [1 / m]]
C = [[1, 0]]
sys = sig.StateSpace(A, B, C, 0)

# Different initial conditions
x_init_1 = [0, 0]
x_init_2 = [1, 0]
x_init_3 = [-1, 0]

# Solve the system for each initial condition
t, y_1, x_1 = sig.lsim(sys, Ft, t, X0=x_init_1)
t, y_2, x_2 = sig.lsim(sys, Ft, t, X0=x_init_2)
t, y_3, x_3 = sig.lsim(sys, Ft, t, X0=x_init_3)

x1_1 = x_1[:, 0]
x2_1 = x_1[:, 1]
x1_2 = x_2[:, 0]
x2_2 = x_2[:, 1]
x1_3 = x_3[:, 0]
x2_3 = x_3[:, 1]

# Plot the Results
plt.figure(1)

plt.subplot(311)
plt.plot(t, x1_1)
plt.title('Simulation of Mass-Spring-Damper System (Initial Condition 1)')
plt.xlabel('t')
plt.ylabel('x1(t)')
plt.grid()

plt.subplot(312)
plt.plot(t, x1_2)
plt.title('Simulation of Mass-Spring-Damper System (Initial Condition 2)')
plt.xlabel('t')
plt.ylabel('x1(t)')
plt.grid()

plt.subplot(313)
plt.plot(t, x1_3)
plt.title('Simulation of Mass-Spring-Damper System (Initial Condition 3)')
plt.xlabel('t')
plt.ylabel('x1(t)')
plt.grid()

plt.tight_layout()
plt.show()

plt.figure(2)

plt.subplot(311)
plt.plot(t, x2_1)
plt.title('Simulation of Mass-Spring-Damper System (Initial Condition 1)')
plt.xlabel('t')
plt.ylabel('x2(t)')
plt.grid()

plt.subplot(312)
plt.plot(t, x2_2)
plt.title('Simulation of Mass-Spring-Damper System (Initial Condition 2)')
plt.xlabel('t')
plt.ylabel('x2(t)')
plt.grid()

plt.subplot(313)
plt.plot(t, x2_3)
plt.title('Simulation of Mass-Spring-Damper System (Initial Condition 3)')
plt.xlabel('t')
plt.ylabel('x2(t)')
plt.grid()

plt.tight_layout()
plt.show()
```
